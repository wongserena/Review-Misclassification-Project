```python
import numpy as np
import pandas as pd
from transformers import pipeline
from tqdm import tqdm
```

```python
reviews = pd.read_csv("processed_reviews.csv")
```

```python
reviews['text'].iloc[5]
```

```python
import textstat

text = reviews['text'].iloc[5]

score = textstat.flesch_kincaid_grade(text)
print("Estimated grade level:", score)
```

```python
batch_size = 1000
grade_levels = []

for i in tqdm(range(0, len(reviews), batch_size)):
    batch = reviews["text"].iloc[i:i+batch_size]
    batch_grades = [textstat.flesch_kincaid_grade(t) for t in batch]
    grade_levels.extend(batch_grades)

reviews["grade_level"] = grade_levels
```

```python
reviews.to_csv("readability_reviews.csv")
```
